```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
import tkinter
from tkinter import filedialog
import os

```

This code will get the root directory for the files you will analyze.  For SynBot, this should be the "Output" folder.


# PSD95

```python
root = tkinter.Tk()
root.withdraw()

dir = tkinter.filedialog.askdirectory()

dir_list = os.listdir(dir)

dir_array = np.array(dir_list)
#print(dir_array)

coloc_list = dir_array[np.char.endswith(dir_array,'greenResults_0.csv')]
coloc_list = coloc_list[np.char.startswith(coloc_list,'img')]
#coloc_list = coloc_list[np.char.startswith(coloc_list,"MAX_")] #May need to change or omit this line depending on the naming of the images.

print(coloc_list)
```

This code will be used to iterate through the csv files and generate a dataframe of frequency counts that has been decoded for genotype and animal. 

```python
n=0
for i in coloc_list:
    current_filepath = dir+"/"+i
    #print(current_filepath)
    current_csv = pd.read_csv(current_filepath)
    
    decode = pd.read_csv(dir+"/"+"Decoder.csv") 
    decode_dict = {}
    rang_d = range(decode.new_names.size)
    keys = ["old_names","Animal","Sex","Genotype","Age"]
    for y in rang_d:
        decode_dict[decode.new_names[y]] = [decode.old_names[y], decode.Animal[y], decode.Sex[y], decode.Genotype[y], decode.Age[y]]
    
    #print(current_csv)
    #Image_ID = [i[4:-19]+".tif" for m in range(0, current_csv.shape[0])]
    #print(Image_ID)
    y_val = np.array(current_csv.Y[:])#If this is for circle based, use this line.
    #y_val = np.array(current_csv.Y[:])#If this is for pixel based, use this line. 
    
    Image_ID = i[0:-19]+".tif" #use this line if going straight from synbot
    #Image_ID = i[4:-17]+".tif" #use this line if images have "MAX_" before them due to separate image preprocessing.
    
    current_hist = plt.hist(y_val, bins =range(8,8200,3908), color = "white")
    plt.close()
    freq_y = current_hist[0][:]
    bins_x = current_hist[1][:current_hist[1].size-1]
    Image_ID_hist = np.array([i[0:-19]+".tif" for m in range(0, bins_x.size)]) #use this line if going straight from synbot
    #Image_ID_hist = np.array([i[4:-19]+".tif" for m in range(0, bins_x.size)]) #use this line if images have "MAX_" before them due to separate image preprocessing.
    
    decode_data = decode_dict[Image_ID]
    decode_df_initial = pd.DataFrame(decode_data, index=keys).T
    decode_df_final = pd.concat([decode_df_initial]*bins_x.size, ignore_index = True)
    
    #line_plot = sb.lineplot(x=bins_x, y=freq_y, color = "black")
    #print("bins",bins_x.size)
    #print(Image_ID_hist)
    current_df = pd.DataFrame(Image_ID_hist,columns=["ImageID"])
    current_df["Green_Counts"]= freq_y
    current_df["Position from pia (pxls)"]= bins_x
    
    current_df = pd.concat([current_df,decode_df_final],axis=1)
    #print(current_df)
    if n == 0:
        syn_counts_df = current_df
        n = n+1
        #print(n)
    else:
        temp_list = [syn_counts_df,current_df]
        syn_counts_df = pd.concat(temp_list)
        n = n+1
        #print(n)

#Exports the data as the specified file name. The first line is the histogram counts for each image.
#The second and third lines are the average counts at each postion for each animal.
#All data should be decoded using the decoding file named "Decoder.csv".
syn_counts_df.to_csv(dir+"/"+"Green_Counts_By500Microns_Based_Unblinded_UppervsLower.csv")
syn_counts_org_df = syn_counts_df.groupby(["Genotype","Animal","Position from pia (pxls)"]).mean(["Green_Counts"])
syn_counts_org_df.to_csv(dir+"/"+"Green_Counts_By500Microns_Based_Unblinded_Organized_UppervsLower.csv")


#Plot the data comparing WT vs KO using the animal averages at each point as the 'n' for SEM.
fig, ax = plt.subplots()
plot_final = sb.lineplot(data=syn_counts_org_df,x="Position from pia (pxls)",y="Green_Counts", 
                         hue = "Genotype", palette = ["red","black"])
ax.set_xlim(-100,8500)
plt.title("PSD95 Counts in V1", fontsize = 15)
plt.suptitle("TLR4 WT vs. KO", fontsize = 17, y = 1.02, fontweight = "bold")
plt.xlabel("Position from pia (pixels)",fontsize = 15)
plt.ylabel("Green_Counts", fontsize = 15)
handles, labels = plt.gca().get_legend_handles_labels()
order = [ 1, 0]
plt.legend([handles[z] for z in order], [labels[z] for z in order])
plt.show()
        
```
